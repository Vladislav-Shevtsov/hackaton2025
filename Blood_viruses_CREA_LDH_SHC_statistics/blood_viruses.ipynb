```python
%matplotlib inline
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import pandas as pd
import time
import scipy as sp
import re
```

```python
X = pd.read_csv('/Users/ksenia/Downloads/blood_viruses.csv', index_col=0)
```

```python
def parse_age(text):
    if pd.isna(text):
        return np.nan

    s = str(text).lower()
    s = re.sub(r'[\u00A0\.]', ' ', s)

    y = re.search(r'(\d+)\s*(?:год|года|лет)\b', s)

    if not y:
        first_num = re.search(r'(\d+)', s)
        return int(first_num.group(1)) if first_num else np.nan

    years = int(y.group(1)) if y else 0

    return years



X['Возраст_в_годах'] = X['Возр.'].apply(parse_age)
```

```python
Y = X[(X['Синоним'] == 'EBV, CMV, HHV6')&(X['Патология'] == 1.0) &(X['Возраст_в_годах'] >= 18)]
```

```python
def parse_viruses(text):
    s = str(text).lower()
    s = re.sub(r'[\u00A0]', ' ', s)
    print(s)

    ebv = re.search(r'количество копий днк ebv на 10.5 клеток\s*\:\s*(\d+)\.00\s*\[>', s)
    if not ebv:
        ebv = re.search(r'вирусная нагрузка ebv\s*\:\s*(\d+)\.00\s*\[>', s)
    ebv_num = int(ebv.group(1)) if ebv else 0
    hhv = re.search(r'количество копий днк hhv6 на 10.5 клеток\s*\:\s*(\d+)\.00\s*\[>', s)
    if not hhv:
        hhv = re.search(r'вирусная нагрузка hhv6\s*\:\s*(\d+)\.00\s*\[>', s)
    hhv_num = int(hhv.group(1)) if hhv else 0
    not_valid = re.search(r'результат исследования невалиден', s)

    return pd.Series({'EBV': ebv_num, 'HHV6': hhv_num, 'Valid': not bool(not_valid)})



Y[['EBV', 'HHV6', 'Valid']] = Y['Результаты'].apply(parse_viruses)

print(Y)
```

```python
Z = X[~(X['Возраст_в_годах'] < 18)]
Z = Z.iloc[:-1]
Z = Z[~(Z['Синоним'] == 'EBV, CMV, HHV6')]
```

```python
for index, row in Y.iterrows():
    if row['Valid'] is True:
        if row['EBV'] > 0:
            new_row = row.copy()
            new_row['Синоним'] = 'эпштейна-барр вирус ПЦР'
            new_row['Результаты'] = 'Эпштейна-Барр (EBV): обнаружен'
            Z = pd.concat([Z, pd.DataFrame([new_row])], ignore_index=True)
        if row['HHV6'] > 0:
            new_row = row.copy()
            new_row['Синоним'] = 'герпес 6 типа ПЦР'
            new_row['Результаты'] = 'герпес тип 6 (HHV6): обнаружен'
            Z = pd.concat([Z, pd.DataFrame([new_row])], ignore_index=True)
```
