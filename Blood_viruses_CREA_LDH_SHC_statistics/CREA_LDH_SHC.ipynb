```python
def parse_age(text):
    if pd.isna(text):
        return np.nan

    s = str(text).lower()
    s = re.sub(r'[\u00A0\.]', ' ', s)

    y = re.search(r'(\d+)\s*(?:год|года|лет)\b', s)

    if not y:
        first_num = re.search(r'(\d+)', s)
        return int(first_num.group(1)) if first_num else np.nan

    years = int(y.group(1)) if y else 0

    return years
```

```python
from datetime import timedelta
```

```python
opp = []
for p in patients: 
    df_p = crea_filtered[crea_filtered['id'] == p]
    df_p_bef = df_p[(df_p['Дата забора'] + timedelta(days=7) >= df_p['date_hsct']) & (df_p['Дата забора'] < df_p['date_hsct'])]
    df_p_aft = df_p[
        (df_p['Дата забора'] >= df_p['date_hsct']) &
        (df_p['Дата забора'] <= df_p['date_hsct'] + timedelta(days=250))
    ]
    if len(df_p_bef) > 0 and  len(df_p_aft) > 0 and max(df_p_bef['Результаты_x']) * 1.5 <= max(df_p_aft['Результаты_x']):
        idx_max_bef = df_p_bef['Результаты_x'].idxmax()
        date_max_bef = df_p_bef.loc[idx_max_bef, 'Дата забора']
        idx_max_aft = df_p_aft['Результаты_x'].idxmax()
        date_max_aft = df_p_aft.loc[idx_max_aft, 'Дата забора']
        opp.append([p, max(df_p_bef['Результаты_x']), date_max_bef, max(df_p_aft['Результаты_x']), date_max_aft, df_p_aft.loc[idx_max_aft, 'date_hsct']])
opp
```

```python
opp = []
for p in patients:
    df_p = crea_filtered[crea_filtered['id'] == p].sort_values('Дата забора')
    df_p_bef = df_p[(df_p['Дата забора'] + timedelta(days=7) >= df_p['date_hsct']) & (df_p['Дата забора'] < df_p['date_hsct'])]
    if df_p_bef.empty:
        continue
    threshold = max(df_p_bef['Результаты_x']) * 1.5
    df_p_aft = df_p[
        (df_p['Дата забора'] >= df_p['date_hsct']) &
        (df_p['Дата забора'] <= df_p['date_hsct'] + timedelta(days=250))
    ]
    if df_p_aft.empty:
        continue
    df_p_aft_above = df_p_aft[df_p_aft['Результаты_x'] >= threshold]
    if df_p_aft_above.empty:
        continue
    first_above = df_p_aft_above.iloc[0]
    opp.append([
        p,
        max(df_p_bef['Результаты_x']),
        df_p_bef.loc[df_p_bef['Результаты_x'].idxmax(), 'Дата забора'],
        first_above['Результаты_x'],
        first_above['Дата забора'],
        first_above['date_hsct']
    ])
```
